```python
#import libraries
import pandas as pd
import numpy as np
import random
```

```python
#reading files
df=pd.read_csv("Desktop/Book.txt")
rdf=pd.read_csv("Desktop/Book_rating.txt",delimiter=r"\s+")

#reading the book names
book_names=np.array(df)[:,1]
```

```python
#calculating singular value decomposition
def SVD(A,dim=2):
    
    #calculating U
    temp=np.dot(A,A.T)
    ueigen,U=np.linalg.eig(temp)
    sortindex=ueigen.argsort()[::-1]
    U=U[:,sortindex]
    U=U[:,:dim]
    U=np.around(U,decimals=2)
    
    #calculating S
    S=np.sqrt(ueigen)
    S=S[:dim]
    S=np.around(S,decimals=2)
    S=np.diag(S)
    
    #calculating V
    temp=np.dot(A.T,A)
    veigen,V=np.linalg.eig(temp)
    sortindex=veigen.argsort()[::-1]
    V=V[:,sortindex]
    V=V[:,:dim]
    V=np.around(V,decimals=2)
    
    return U,S,V
 
    
#splitting the dataset into train and test set    
def split(M,train,test): 
    for i in range(np.shape(M)[0]):
        if random.random()<0.70:
            train.append(M[i])
        else:
            test.append(M[i])
            
```

```python
M=np.array(rdf)
train=[]
test=[]

#splitting
split(M,train,test)

train=np.array(train)
test=np.array(test)

train1=train[:,1:]
test1=test[:,1:]

train1=np.abs(train1.astype(int))
test1=np.abs(test1.astype(int))
#svd
U,S,V=SVD(train1)
U=U.astype(float)
V=V.astype(float)
S=S.astype(float)
```

```python
#predicting ratings
for i in range(np.shape(test1)[0]):
    prd=np.dot(test1[i],V)
    Vt=np.transpose(V)
    result=np.dot(prd,Vt)
    result=np.array(result)
    
    #sorting the results
    sortindex=result.argsort()
    sortindex=sortindex[::-1]
    res=book_names[sortindex]
    
    #printing the first 10 recommended books for the user
    print("\nRecommendation list for " +test[i][0] +" :\n")
    for i in range(10):
        print(res[i])
        
    
```
